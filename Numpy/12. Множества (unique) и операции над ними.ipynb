```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 4, 4, 3, 2, 1])
```

```python
setA = np.unique(a)
print(setA)
```

```python
print(np.unique(a, return_counts = True)) #также возвращает число вхождений элементов в исходном массиве a.
```

```python
print(np.unique(a, return_index = True)) # позволяет определять индексы первого вхождения уникальных элементов
```

```python
print(np.unique(a, return_inverse = True)) # возвращает индексы
```

```python
setA, indx = np.unique(a, return_inverse = True)
print(setA)
print(indx)
```

```python
print(setA[indx])
```

```python
x = np.array([[0, 1, 1, 2],[0, 1, 1, 2],[9, 1, 1, 2]])
print(x)
print(np.unique(x)) # array([0, 1, 2, 9])
```

```python
print(np.unique(x, axis = 0)) # возвращает уникальные строки 
```

```python
print(np.unique(x, axis = 1)) # возвращает уникальные столбцы
```

## Операции над множествами

```python
x = np.array([0, 1, 2, 3])
y = np.array([1, 2, 3, 4, 5, 6, 7, 8])
print(x)
print(y)
```

```python
print(np.in1d(x, y)) # показывает, есть ли элемент массива x в массиве y
```

```python
print(np.intersect1d(x, y)) # пересечение множеств
```

```python
print(np.union1d(x, y)) # объединение множеств
```

```python
print(np.setdiff1d(x, y)) # вычитание множеств
print(np.setdiff1d(y, x))
```

```python
print(np.setxor1d(x, y)) # остаются не совпадающие значения из двух множеств.
```
