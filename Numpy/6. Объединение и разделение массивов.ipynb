```python
import numpy as np
#Функции hstack и vstack
```

```python
a = np.array([(1, 2), (3, 4)])
b = np.array([(5, 6), (7, 8)])
```

```python
print(a)
print(b)
```

```python
print(np.hstack([a, b])) # объединение по горизонтали, по axis1
print(np.vstack([a,b])) # по вертикали, по строкам, по axis0
```

```python
a = np.arange(1, 13)
b = np.arange(13, 26)
a.resize(3, 3, 2)
b.resize(3, 3, 2)
```

```python
print(a)
print(b)
```

```python
c0 = np.concatenate([a, b], axis=0) # размерность 6x3x2
c1 = np.concatenate([a, b], axis=1) # размерность 3x6x2
c2 = np.concatenate([a, b], axis=2) # размерность 3x3x4
```

```python
print(c0)
```

```python
print(c1)
```

```python
print(c2)
```

```python
print(np.r_[ [1, 2, 3], 4, 5])  # список + дополнительные элементы
```

```python
print(np.r_[ 1:9, 90, 100]) # срез + два элемента
```

```python
print(np.r_[ np.array([1,2,3]), np.array([4,5,6])]) # объединение двух массивов
```

```python
print(np.r_[ [(1,2,3), (4,5,6)], [(7,8,9)] ]) # объединение двумерного и одномерного списков
```

```python
np.c_[1:5] # объединение выполняется по второй оси axis1
```

```python
np.c_[ [1, 2, 3], [4, 5, 6]]
```

```python
np.c_[ [(1,2,3), (4,5,6)], [[7],[8]] ]
```
