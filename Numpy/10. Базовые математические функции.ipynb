## Функции sum, mean, min и max

```python
import numpy as np
```

```python
a = np.array([ 1,  2,  3, 10, 20, 30])
print(a)
```

```python
print(a.sum())
```

```python
print(a.mean())
print(a.max())
print(a.min())
```

```python
a.resize(3, 2)
print(a)
```

```python
print(a.sum(axis = 0))
print(a.sum(axis = 1))
```

```python
print(a.max(axis = 0))
print(a.max(axis = 1))
```

## Тригонометрические функции

```python
a = np.linspace(0, np.pi, 10)
print(a)
```

```python
res1 = np.sin(a) # возвращает массив синусов углов
print(res1)
res2 = np.cos(a) # возвращает массив косинусов углов
print(res2)
print(np.arcsin(res1)) # возвращает арксинусы от значений res1
```

```python
print(np.sin(np.pi/3))
print(np.cos([0, 1.57, 3.17]))
```

## Функции генерации псевдослучайных чисел

```python
#Функции rand() и randint() генерируют числа с равномерным законом распределения
```

```python
print(np.random.rand()) # вещественное случайное число от 0 до 1
```

```python
print(np.random.rand(5)) # массив таких чисел
```

```python
print(np.random.rand(2, 3)) # Для получения двумерных массивов – два аргумента
```

```python
print(np.random.randint(10)) # генерация целых чисел в диапазоне [0; 10)
```

```python
print(np.random.randint(5, 10)) # генерация в диапазоне [5; 10)
```

```python
#Для получения массива из таких случайных чисел дополнительно следует указать параметр size, следующим образом:
```

```python
print(np.random.randint(5, size = 4))
```

```python
print(np.random.randint(1, 10, size = (2, 5)))
```

```python
print(np.random.randn()) # нормальная СВ с нулевым средним и единичной дисперсией
```

```python
print(np.random.randn(5)) # массив из пяти нормальных СВ
```

```python
print(np.random.randn(2, 3)) # матрица 2x3 из нормальных СВ
```

По умолчанию, все рассмотренные функции при каждом новом запуске генерируют разные числа. Однако, иногда требуются одни и те же их реализации. Например, такой подход часто используется для сохранения начальных случайных значений весов в нейронных сетях. Для этого устанавливают некое начальное значение зерна (seed):

```python
np.random.seed(1) # начальное значение генератора случайных чисел
np.random.randint(10, size=10)
```

```python
np.random.randint(10, size=10)
```

```python
np.random.seed(1) 
```

```python
np.random.randint(10, size=10)
```

## Функции перемешивания элементов массива

```python
a = np.arange(10)
print(a)
```

```python
np.random.shuffle(a) # перемешивает случайным образом элементы массива. Работает только с первой осью axis0
print(a)
```

```python
a = np.arange(1, 10).reshape(3, 3)
print(a)
```

```python
np.random.shuffle(a) # перемешивает только строки
print(a)
```

```python
print(np.random.permutation(10)) # возвращает случайную последовательность чисел, генерируя последовательность «на лету»
```

## Функции математической статистики

```python
x = np.array([1, 4, 3, 7, 10, 8, 14, 21, 20, 23])
y = np.array([4, 1, 6, 9, 13, 11, 16, 19, 15, 22])
```

```python
print(np.median(x))
```

```python
print(np.var(x)) # дисперсия
```

```python
print(np.std(x)) # среднеквадратическое отклонение
```

```python
XY = np.vstack([x, y])
print(XY)
```

```python
print(np.corrcoef(XY)) #коэффициент корреляции Пирсона
```

```python
print(np.cov(XY)) # вычисление ковариационной матрицы
```

```python
print(np.correlate(x, y))# вычисление кросс-корреляции
```
