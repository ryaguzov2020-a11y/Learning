```python
import numpy as np
```

```python
a = np.array([1,2,3,4])
print(a)
print(type(a))
```

```python
a = np.array([1,2,3,4], 'float') # можно указать поддерживаемый тип данных
print(a)
```

```python
print(np.complex64(10))
```

```python
a = np.array([1,2,5000,1000])
print(a)
b = np.complex64(a)
print(b)
c = np.int32(b)
print(c)
```
