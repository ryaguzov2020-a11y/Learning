## Индексация и срезы одномерных массивов

```python
import numpy as np
```

```python
a = np.arange(12)
print(a)
```

```python
print(a[2])
print(a[-1])
print(a[0])
```

```python
a[0] = 100
print(a)
```

```python
b = a[2:4]
print(b)
```

```python
b[0] = -100
print(b)
print(a)
```

```python
print(a[3:])
print(a[5:])
print(a[:])
print(a[1:6:2])
```

```python
a[:4] = [-1, -2, -3,- 4]
print(a)
```

```python
a[4::2] = np.array([10, 20, 30, 40])
print(a)
```

```python
for x in a:
    print(x, sep= ' ', end=' ')
```

## Индексация и срезы многомерных массивов

```python
x = np.array([(1, 2, 3), (10, 20, 30), (100, 200, 300)])
print(x)
```

```python
print(x[1, 1])
print(x[-1, -1])
print(x[0])
print(x[0, :])
print(x[0, :2])
print(x[:, 1])
```

```python
for row in x:
    for val in row:
        print(val, end=' ')
    print()
```

```python
for val in x.flat:
    print(val, end= " ")
```

```python
a = np.arange(1, 82).reshape(3, 3, 3, 3)
print(a[1, 2, 0, 1])
```

```python
print(a[:, 1, :, :])
```

```python
print(a[:, :, 0, 0])
```

```python
print(a[0:2, 0:2, 1, 1]) # матрица 2x2
```

# Списочная индексация

```python
a = np.arange(1, 9)
print(a)
```

```python
b = a[[0]] # копия
print(b)
```

```python
b = a[[0, 1, 7, 5]] # возвращает элементы с указанными индексами
print(b)
```

```python
b = a[[0, 0, 1, 1, 1, 2, 3, 4, 5, 6, 7]]
print(b)
```

```python
a[[0, 1, 7, 5]] = [10, 20, 30, 40] # в элементы массива a с индексами 0,1 и т.д. заносятся значения 10, 20 и т.д.
print(a)
```

```python
indx = [0, 0, 1, 1, 1, 2] # это индексы
print(a[indx])
```

```python
bIndx = [True, True, False, False, False, True, False, False] # где True, возвращает значение  
print(a[bIndx])
```

```python
i = a > 5
print(i)
print(a[i])
```

```python
print(a[a > 5])
```

## Списочная индексация и многомерные массивы

```python
a = np.arange(1, 9)
i = np.array([[0, 1], [2, 3]]) # тоже индексы
print(a[i])
```

```python
a = np.arange(1, 13).reshape(3, 4)
print(a)
```

```python
indx = [2, 1, 0] # здесь индексы обозначают номера строк двумерного массива
print(a[indx])
```

```python
indx = np.array([[1, 0], [2, 1]]) # тоже порядок строк
print(a[indx])
```

```python
i0 = [0, 1]
i1 = [1, 2]
print(a[i0, i1])
```

```python
print(a[:, [1, 2]])
```

```python
print(a[[0, 1], 1])
```

## Изменение массивов через списочную индексацию

```python
a = np.arange(7)
print(a)
```

```python
a[[0, 4, 6]] = [-1, -2, -3]
print(a)
```

```python
a[[0, 1, 2]] += 1
print(a)
```

```python
print(a[[0, 1, 2]])
```
