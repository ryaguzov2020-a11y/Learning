```python
import numpy as np
```

## Матричное умножение

```python
a = np.arange(1, 10).reshape(3, 3)
b = np.arange(10, 19).reshape(3, 3)
print(a)
print(b)
```

```python
print(np.dot(a, b)) # каждая строка умножается на каждый столбец
```

```python
print(np.dot(b, a))
```

```python
print(np.matmul(a, b)) # тот же результат
```

## Векторное умножение 

```python
a = np.arange(1, 10)
b = np.ones(9)
print(a)
print(b)
```

```python
print(np.dot(a, b))
```

```python
print(np.inner(a, b)) # внутреннее умножение (первый вариант)
```

```python
print(np.outer(a, b)) # внешнее умножение (второй вариант)
```

```python
print(a @ b) # перегруженный оператор, заменяющий функцию matmul и dot
```

```python
a.resize(3, 3)
b.resize(3, 3)
print(a)
print(b)
```

```python
print(a @ b)
```

## Умножение вектора на матрицу

```python
a = np.array([1,2,3])
b = np.arange(4,10).reshape(3,2)
print(a)
print(b)
```

```python
print(np.dot(a, b)) # вектор-строка умножается на каждый столбец матрицы
```

```python
a = np.array([1, 2])
print(np.dot(b, a)) # вектор-столбец умножается на каждую строку матрицы
```

```python
print(b @ a)
```

## Элементы линейной алгебры

```python
a = np.array([(1, 2, 3), (1, 4, 9), (1, 8, 27)])
print(a)
```

```python
print(np.linalg.matrix_rank(a)) # возвращает ранг матрицы
```

```python
y = np.array([10, 20, 30]) 
print(y)
```

```python
print(np.linalg.solve(a, y)) # возвращает корни уравнения
```

```python
invA = np.linalg.inv(a) # возвращает обратную матрицу
print(invA)
```

```python
print(a @ invA) #единичная матрица
```

```python
print(invA @ y) # возвращает корни уравнения
```
