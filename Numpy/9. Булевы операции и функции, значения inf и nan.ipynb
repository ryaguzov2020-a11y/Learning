## Функции greater, less, equal

```python
#greater(a, b) – выполняет сравнение a > b;
#less(a, b) – выполняет сравнение a < b;
#equal(a, b) – выполняет сравнение a == b.
```

```python
print(np.greater(a, b))
```

```python
print(np.less(a, b))
```

```python
print(np.equal(a, b))
```

## Функции array_equal, all и any

```python
print(np.array_equal(a, b))
```

```python
if np.array_equal(a, b):
    print("a == b")
```

```python
print(np.any(a > 5)) # хотя бы один элемент больше 5
```

```python
print(np.any(a == 5))
```

```python
print(np.all(a > 5)) # все элементы массива а больше 5
```

## Функции isnan и isinf

```python
b = np.array([1, 2, np.nan, np.inf, -np.inf])
print(b)
```

```python
print(np.isinf(b))
print(np.isnan(b))
```

```python
#можно исключить нечисловые элементы из массива
indx = np.isinf(b)
print(b[~indx])  #  ~indx инвертирует булевы значения
```

## Функции logical_and, logical_or, logical_not и logical_xor


```python
X = np.array([True, False, True, False])
Y = np.array([True, True, False, False])
```

```python
print(np.logical_and(X, Y)) # True, если принимают одинаковые значения 
```

```python
print(np.logical_or(X, Y)) # True, если принимают разные значения 
```

```python
print(np.logical_not(X)) # инвертирует 
```

```python
print(np.logical_xor(X, Y))
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 3, 2, 1])
```

```python
print(np.logical_and(x > 2, y > 2))
```

```python
print(np.logical_or(x > 2, y > 2))
```
