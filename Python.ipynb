# List comprehensions(генератор списков)

```python
nums = [1, 2, 3, 4, 5]
```

```python
new = []
for num in nums:
    new.append(num * 2)
new
```

```python
new = [num * 2 for num in nums]
new
```

```python
new = [num * 2 for num in nums if num > 2]
new
```

```python
new = []
for num in nums:
    if num > 2:
        new.append(True)
    else:
        new.append(False)
new
```

```python
new = [True if num > 2 else False for num in nums]
new
```

```python
[[1 for i in range(3)] for j in range(3)] # вложенный генератор списков
```

```python
A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

AA = [[num ** 2 for num in row] # выполняется вторым
      for row in A ] # выполняется первым
print(AA)
```

# Классы

```python
it_specialists = [('Jon', 'DE', 150), 
                  ('Carl', 'DA', 120), 
                  ('Gleb', 'DS', 300)
                 ]
```

```python
class ItSpecialists: # класс
    # конструктор класса
    def __init__(self, name, speciality, salary): # атрибуты объекта, self - параметр ссылка на текущий объект
        self.name = name
        self.speciality = speciality
        self.salary = salary
    def tell_about_yourself(self): # метод класса
        return f'Hi! My name is {self.name}. I\'m a {self.speciality}. My salary is {self.salary}'

#Методы класса фактически представляют функции, которые определенны внутри класса и которые определяют его поведение.
```

```python
jon = ItSpecialists('Jon', 'DE', 150) # объект класса
jon
```

```python
jon.name
```

```python
carl = ItSpecialists('Carl', 'DA', 120)
carl
```

```python
jon.tell_about_yourself()
```

```python
carl.tell_about_yourself()
```

# Наследование классов

```python
class DataScientist(ItSpecialists):

    def __init__(self, name, salary):
        super().__init__(name, 'DS', salary) # super - вызвать методы верхнего класса.

    def tell_about_yourself(self): # метод класса
        return f'Hi! My name is {self.name}. I\'m a {self.speciality}. My salary is {self.salary}k per second'
    
    def ml(self, data):
        return 'profit for company and myself'
```

```python
gleb = DataScientist('Gleb', 300)
```

```python
gleb.tell_about_yourself()
```

```python
gleb.ml(123)
```
