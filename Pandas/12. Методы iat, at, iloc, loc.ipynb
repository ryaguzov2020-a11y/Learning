```python
import numpy as np
import pandas as pd
```

```python
dict_array = {'age': [53, np.nan, 11, 18, 7],
              'name': ['Сергей', 'Маша', 'Ксюша', 'Аристарх', 'Соня'],
              'наличие авто': [True, True, False, True, False]
             }

df = pd.DataFrame(dict_array,
                 index = ['row_1', 'row_2', 'row_3', 'row_4', 'row_5'])

df
```

## Доступ к одному элементу

```python
# синтаксис df.iat['индекс строки', 'индекс столбца']
df.iat[2, 1]
```

```python
df['age'].iat[3]
```

```python
df.iat[4, 0] = 22 # замена
df
```

```python
# синтаксис df.at['метка строки', 'метка столбца']
# то же самое только метки вместо индексов
df.at['row_3', 'name']
```

```python
df.at['row_4', 'age'] = 35
df
```

```python
# поменять имена столбцов
df.loc[:, ['age', 'name']] = df[['name', 'age']].to_numpy()
```

```python
df.loc['row_1':'row_3', ['age', 'name']] = pd.DataFrame({ 'age': [22, 18, 17],
                                                          'name': ['Сергей', 'Маша', 'Даша'] },
                                                          index=['row_1', 'row_2', 'row_3']
                                                       )

df
```

```python
df.loc['row_4':'row_5', 'age'] = pd.Series([20, 12],
                                          index=['row_4', 'row_5'])
df
```
