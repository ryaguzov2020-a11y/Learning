```python
import numpy as np
import pandas as pd
```

## Метод Rolling

```python
df_date = pd.DataFrame({'col_1': [0, 1, 2, 3, 4],
                       'col_2': [90, 91, 92, 93, 94]}
                      )
df_date
```

```python
a = df_date['col_2'].rolling(3).mean()
a
```

```python
df_date.rolling(3)
for window in df_date.rolling(3):
    print(window, end ='\n\n')
```

```python
df_date.rolling(3).sum()
```

```python
print(df_date.rolling(3).mean(), end='\n\n')
print(df_date.rolling(3, min_periods=1).mean())
```

```python
print(df_date.rolling(3).mean(), end='\n\n')
print(df_date.rolling(3, step=2).mean())
```

```python
df_date.rolling(2).agg({'col_1': 'sum', 'col_2': 'mean'})
```

```python
df = pd.DataFrame({'sales': np.random.randint(0, 50, 100)},
                 index=pd.date_range('2023', periods=100))
df.head(10)
```

```python
df_rolling = df.rolling(7, min_periods=1).agg(['sum', 'mean', 'max', 'min'])
df_rolling.head(10)
```

```python
df_rolling.reset_index(inplace=True)
df_rolling
```

```python
df_rolling['weekday'] = df_rolling['index'].dt.dayofweek
df_rolling
```

```python
df_rolling.loc[df_rolling['weekday'] == 6]
```
