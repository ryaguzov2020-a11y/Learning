```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.DataFrame({'col_1':[5, 6, np.nan, 3, 7],
                     'col_2':[27, 6, 7, 40, 70],
                     'col_3':[1, 45, 6, 62, 7]},
                   index = ['row_1', 'row_2', 'row_3', 'row_4', 'row_5']
                   )
df_1
```

```python
df_2 = pd.DataFrame({'col_1':[5, 6, np.nan, 3, 7],
                     'col_2':[27, 6, 7, 40, 70],
                     'col_3':[1, 45, 6, 62, 7]},
                   index = ['row_1', 'row_2', 'row_3', 'row_4', 'row_5']
                   )
df_2
```

```python
series = pd.Series([5, 6, np.nan, 3, 7],
                  index=['row_10', 'col_2', 'row_3', 'row_4', 'row_5']
                  )
series
```
