```python
import pandas as pd
```

## Функция concat()

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index= [0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index= [1, 2, 3, 4])
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                    index= [2, 3, 4, 5])
```

```python
print(df1, end='\n\n')
print(df2, end='\n\n')
print(df3)
```

```python
pd.concat([df1, df2, df3])
```

```python
pd.concat([df1, df2, df3], ignore_index=True)
```

```python
pd.concat([df1, df2, df3], axis = 1)
```

## Метод merge

```python
df_1 = pd.DataFrame({'client_id': [100103, 21990, 455323, 1001003, 21990, 455323, 455323, 21990, 455323],
                    'product': ['product_2', 'product_2', 'product_1', 'product_1', 'product_1', 'product_3','product_3', 'product_3', 'product_6']})
```

```python
df_2 = pd.DataFrame({'product': ['product_1', 'product_2', 'product_3', 'product_4', 'product_5'],
                     'price': [1000, 2000, 3000, 4000, 5000]})
```

```python
print(df_1, end='\n\n')
print(df_2)
```

```python
df_1.merge(df_2, how = 'left') # элемент левого дф ищется в правом дф и возвращается
```

```python
df_1.merge(df_2, how = 'right') # элемент правого дф ищется в левом дф и возвращается
```

```python
df_1.merge(df_2, how = 'inner') # эл-ты которые присутствуют в обоих дф
```

```python
df_1.merge(df_2, how = 'outer') # все элементы
```

```python
df_1.merge(df_2, how = 'inner', on = 'product')
```

```python
df_1 = pd.DataFrame({'client_id': [100103, 21990, 455323, 1001003, 21990, 455323, 455323, 21990, 455323],
                    'product': ['product_2', 'product_2', 'product_1', 'product_1', 'product_1', 'product_3','product_3', 'product_3', 'product_6'],
                    'color_prod': ['black', 'white', 'white', 'black', 'white', 'white', 'black', 'white', 'black']}
                   )
```

```python
df_2 = pd.DataFrame({'product': ['product_1', 'product_2', 'product_3', 'product_4', 'product_5', 'product_1', 'product_2', 'product_3', 'product_4', 'product_5'],
                     'price': [1000, 2000, 3000, 4000, 5000, 1500, 2500, 3500, 4500, 5500],
                     'color_prod': ['black', 'black', 'white', 'white', 'black', 'white', 'white', 'white', 'black', 'white']})
```

```python
df_1
```

```python
df_2
```

```python
df_1.merge(df_2, how='inner', on='product')
```

```python
df_1.merge(df_2, how='inner', on=['product', 'color_prod'])
```

```python
df_1.rename(columns={'product': 'prod'}, inplace=True)
df_2.rename(columns={'color_prod': 'color'}, inplace=True)
```

```python
df_1
```

```python
df_2
```

```python
df_1.merge(df_2, how='left', 
           left_on=['prod', 'color_prod'],
           right_on=['product', 'color'])
```

```python
df_1 = pd.DataFrame({'product': ['product_1', 'product_2', 'product_3', 'product_4', 'product_5'],
                     'count': [10, 20, 30, 40, 50]})
```

```python
df_2 = pd.DataFrame({'product': ['product_3', 'product_4', 'product_5', 'product_6', 'product_7'],
                     'count': [100, 200, 300, 400, 500]})
```

```python
df_1.merge(df_2, how = 'outer', on='product')
```

```python
df_1.merge(df_2, 
           how = 'outer', 
           on='product',
          suffixes=['_shop_left', '_shop_right'])
```
