## Уникальные значения и дубликаты в индексах

```python
df_data = pd.DataFrame({'col_1': [1, 4, 3, 5, 6, 3, 7, 13],
                        'col_2': [2, 5, 6, 7, 4, 3, 8, 9]},
                      index = index_data, # назначение имен
                      columns = column_data) # выборка
df_data
```

```python
index_data = df_data.index
print(f'{index_data}')
columns_data = df_data.columns
print(f'{columns_data}')
```

```python
index_data.unique() # уникальные значения индексов
```

```python
columns_data.nunique() # количество уникальных значений 
```

```python
index_data.nunique()
```

```python
columns_data.is_unique # True если все значения уникальны
```

```python
print(index_data)
print(index_data.duplicated()) # False если элемент встречается впервые
```

## Пропуски в индексах

```python
index = pd.Index([1, np.nan, 3, np.nan, 5])
index
```

```python
df_nan = pd.DataFrame(np.arange(5),
                      index = index,
                     columns = ['col_1'])
df_nan
```

```python
df_nan.index.hasnans # проверка наличия пропусков в индексах
```

```python
df_nan.index.isna()
```

```python
df_nan.index.dropna() # удаляет пропуски в индексах, при этом не меняя индексацию
```
