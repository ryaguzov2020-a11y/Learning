```python
import numpy as np
import pandas as pd
```

```python
dict_array = {'age': [53, 53, 11, 18, 18, np.nan],
              'name': ['Сергей', 'Сергей', 'Ксюша', 'Аристарх', 'Соня', 'Соня'],
              'наличие авто': [True, True, False, True, False, True]}
```

```python
df = pd.DataFrame(dict_array, 
                 index = ['row_0', 'row_1', 'row_2', 'row_3', 'row_4', 'row_5']
                 )
df
```

```python
#Метод unique
df['age'].unique()
```

```python
#Метод nunique
df['age'].nunique()
```

```python
df.nunique()
```

```python
df.value_counts() # считает количество встречающихся строк
```

```python
df['наличие авто'].value_counts()
```

```python
df.value_counts(['age', 'name'], dropna=False)
```

```python
df.value_counts(['name'], normalize=True)
```

```python
df.sort_values('age')
```

```python
df.sort_values(['age', 'name'])
```

```python
df.sort_values('age', na_position='first')
```

```python
df.sort_values('age', na_position='first', inplace= True)
df
```

```python
df.sort_values('age', na_position='first', ascending=False, inplace= True) # ascending=False по убыванию
df
```
