```python
import pandas as pd
import numpy as np
```

```python
pd.Series()
```

```python
# Серия на основе списка
s = pd.Series( [5, 6, 9.8, 1, 7],
             index = ['row_1', 'row_2', 'row_3', 'row_4', 'row_5'], # индексы строк
             dtype = 'string', # тип данных
             name = 'pd_Series') # название серии
s
```

```python
#массив данных в виде словаря
dict_array = {'row_1': 2,
             'row_2': 3,
             'row_3': 4,
             'row_4': 5, 
             'row_5': 6}
print(dict_array)
```

```python
# Серия на основе словаря
s = pd.Series( dict_array,
             #index = ['row_3', 'row_4', 'row_100'], # делается выборка по заданным индексам строк
             dtype = 'int', # нет ошибки, т.к. сначала pandas приводит данных к инту, а потом делает выборку
             name = 'pd_Series') # название серии
print(s)
```

```python
# Серия на основе массива numpy
np_array = np.array([5, 6, 6, 8, 7])
s = pd.Series( np_array,
              copy = False) #не копирует данные, а использует данные указанного массива
print(s)
```

```python
#изменение элемента
s.iloc[3] = 111
print(s)
print(np_array)
```

```python
# Серия на основе массива numpy
np_array = np.array([5, 6, 6, 8, 7])
s = pd.Series( np_array,
              copy = True) #не копирует данные, а использует данные указанного массива
print(s)
```

```python
#изменение элемента
s.iloc[3] = 111
print(s)
print(np_array)
```

```python
# Серия на основе серии
s_2 = pd.Series(s,
               index =[1, 2, 4])
print(s_2)
```

## Создание DataFrame

```python
dict_array = {'col_1':[5, 6, 6, 3, 7],
             'col_2':[27, 2, 6, 3, 7],
             'col_3':[5, 6, 3, 3, 4],
             'col_4':[65, 6, 7, 82, 7],
             'col_5':[5, 17, 26, 1, 7]}
```

```python
# DataFrame based on dictionary
df = pd.DataFrame(dict_array, # ключи словаря становятся метками столбцов
                 index= ['row_1', 'row_2', 'row_3', 'row_4', 'row_5'],
                 columns= ['col_4', 'col_1', '5'], #выборка
                 dtype= float )
print(df)
```

```python
np_array = np.array([[5, 6, 6, 3, 7],
             [27, 2, 6, 3, 7],
             [5, 6, 3, 3, 4],
             [65, 6, 7, 82, 7],
             [5, 17, 26, 1, 7]])
```

```python
df = pd.DataFrame(np_array, copy= True)
df
```

```python
df.loc[:, 3:4] = 111
df
```
