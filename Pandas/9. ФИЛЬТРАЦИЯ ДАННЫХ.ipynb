```python
import pandas as pd
import numpy as np
```

```python
dict_array = {'age': [53, 37, 11, 18, 7],
              'name': ['Сергей', 'Маша', 'Ксюша', 'Аристарх', 'Соня'],
              'наличие авто': [True, True, False, True, False]
             }

df = pd.DataFrame(dict_array)

df
```

```python
df[ [True, False, True, False, True] ]
```

```python
df['наличие авто']
```

```python
df[df['наличие авто']]
```

```python
df['age'] <= 11 
```

```python
df[df['age'] <= 11]
```

```python
df[(df['age'] <= 18) & (df['наличие авто'])]
```

```python
df[(df['age'] > 18) & (df['наличие авто'])]
```

```python
df[(df['age'] == 18) | ~df['наличие авто']]
```
