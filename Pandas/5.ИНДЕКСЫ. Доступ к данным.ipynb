```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'col_1':[5, 6, 6, 3, 7],
                  'col_2':[3, 2, 6, 3, 7],
                  'col_3':[5, 6, 3, 3, 4],
                  'col_4':[5, 6, 7, 8, 7],
                  'col_5':[5, 1, 6, 1, 5]}
                 )
df
```

```python
# установка значений столбца в качестве индексов строк
df.set_index('col_1')
```

```python
df.set_index(['col_1', 'col_2'],
             append = True,
             inplace = True
            )
df
```

```python
df.reset_index(['col_2']) # превращает индексы строки в значения столбца
```

```python
df.reset_index(inplace = True)
df
```

## Доступ к данным с помощью индексов строк и столбцов

```python
d = {'col_1':[5, 6, 6, 3, 7],
     'столбец_2':[3, 2, 6, 3, 7],
     'col_3':[5, 6, 3, 3, 4],
     'col_4':[5, 6, 7, 8, 7]
    }
df = pd.DataFrame(d)
df
```

```python
df['col_1']
```

```python
df[['col_1']]
```

```python
#Выводит несколько столбцов
df[['col_1', 'col_4']]
```

```python
# добавляет новый столбец
df['new_col'] = 1
df
```

```python
#создает новый столбец со значениями из указанного столбца
df['new_col_2'] = df['col_3'] + df['col_4']
df
```

```python
df['new_col_3'] = np.arange(5)
df
```

```python
# удаление столбца
del df['new_col_3']
df
```
