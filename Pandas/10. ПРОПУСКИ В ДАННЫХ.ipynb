```python
import pandas as pd
import numpy as np
```

```python
dict_array = {'age': [53, np.nan, 11, 18, 7],
              'name': ['Сергей', 'Маша', 'Ксюша', 'Аристарх', 'Соня'],
              'наличие авто': [True, True, False, np.nan, False],
              'марка авто': [np.nan, np.nan, np.nan, np.nan, np.nan,]
             }

df = pd.DataFrame(dict_array)

df
```

## Метод dropna()

```python
df.dropna()
```

```python
df[['age', 'name', 'наличие авто']].dropna() # удаляет строки, в которых есть пропуски 
```

```python
#df.dropna(axis = 0) # для строк
df.dropna(axis = 1) # для столбцов
```

```python
# Аргумент how
df.dropna(how='all', axis = 1) # удаляет столбцы, в которых ВСЕ элементы - пропуски 
```

```python
df
```

```python
# Аргумент subset
df.dropna(subset=['age', 'наличие авто']) # удаляет те строки, для которых в указанных столбцах есть пропуски
```

```python
df.dropna(subset=[1], axis=1) # удаляет те столбцы, для которых в указанных строках есть пропуски
```

```python
df.dropna(subset=['age', 'наличие авто'],
         inplace = True)

df
```
