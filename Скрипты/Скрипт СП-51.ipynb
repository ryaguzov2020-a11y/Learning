```python
import pandas as pd
```

```python
df= pd.read_excel('~/Desktop/an notebooks/работа 2026/отчет_трен.xlsx')
```

```python
df = df.transpose()
```

```python
df.columns = list(df.iloc[0,:])
```

```python
df.drop('Дата', axis=0, inplace=True)
```

```python
df.head()
```

```python
def fill_column(df):
    for i in df.columns:
        if 'Площадка' in i:
            df[i] = i
    return df
```

```python
fill_column(df)
```

```python
col_start = 0
col_end = 5
dataframes_list = []
for i in range(1, 6):
    df_i = df.iloc[:, col_start:col_end+1]
    for i in df_i.columns:
        if 'Площадка' in i:
            df_i.rename(columns={i: 'Площадка'}, inplace=True)
    dataframes_list.append(df_i)
    col_start += 5
    col_end += 5
```

```python
dataframes_list[0]
```

```python
df_all = pd.concat([dataframes_list[0],
                    dataframes_list[1],
                    dataframes_list[2],
                    dataframes_list[3],
                    dataframes_list[4]])
df_all['Общее_количество_на_дату'] = df_all.iloc[:, 1:].sum(axis=1)
```
