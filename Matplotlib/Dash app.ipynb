# Отображение

```python
# создаем объект класса Dash()
app = Dash()

app.layout = [
    #заголовок первого уровня
    html.H1(children='Title of Dash App', style={'textAlign':'center'}),
    # выпадающий список. multi=True - с множественным выбором. Его опции берутся из уникальных значений столбца country датафрейма df. 
    # Значение по умолчанию – 'пустой список'. У элемента есть id='dropdown-selection' для связи с callback-функциями.
    dcc.Dropdown(options = df.country.unique(), value = [], multi=True, id='dropdown-selection', placeholder='Выберите страну'),
    # выпадающий список. multi=True - с множественным выбором. Его опции берутся из уникальных значений столбца continent датафрейма df. 
    # Значение по умолчанию – 'пустой список'. У элемента есть id='dropdown-continent' для связи с callback-функциями.
    dcc.Dropdown(options = df.continent.unique(), value = [], multi=True, id='dropdown-continent', placeholder='Выберите континент'),
    # пустой контейнер для графика с id='graph-content'
    dcc.Graph(id='graph-content')
]

# callback - для отображения графика
@callback(
    # Результат работы функции (объект графика) отправляется обратно в компонент dcc.Graph, обновляя его содержимое.
    Output('graph-content', 'figure'),
    # На вход - принимает значение из выпадающего списка и передает его в функцию. Связан с dcc.Dropdown по id = 'dropdown-selection'
    Input('dropdown-selection', 'value'),
    Input('dropdown-continent', 'value')
)

def update_graph(countries, continent):
    dff = df[(df.country.isin(countries)) & (df.continent.isin(continent))]
    fig = px.line(dff, x='year', y='pop', color='country', markers='True', text="pop")

    title_text = 'Country population'
    if continent:
        title_text += f', Континент: {", ".join(continent)}'
    if countries:
        title_text += f', Страны: {", ".join(countries)}'

    fig.update_layout(
                      #параметры заголовка
                      title={'text':title_text, 
                             'xanchor': 'right',
                             'yanchor': 'top',
                             'y': 0.95,
                             'x': 0.6,
                             'font':{'size':18}
                            },
                      # шрифт оси x и y
                      font = {'size': 16},
                      # параметры оси x
                      xaxis={
                             'title' :{'text':'Year'}
                            },
                      # параметры оси y
                      yaxis={
                             'title' :{'text':'Population'}
                            }
                        )
    fig.update_traces(textposition="bottom right")

    
    return fig
# конструкция Python для запуска приложения.
# __name__ – специальная переменная Python, которая хранит имя текущего модуля
# __main__ – значение, которое получает __name__, когда файл запущен напрямую (а не импортирован как модуль)
if __name__ == '__main__':
    app.run(debug=True)
```
