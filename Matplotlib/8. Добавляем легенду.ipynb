```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(0, 5, 0.25), '--o', label='line 1')
ax.plot(np.arange(0, 10, 0.5),  ':s', label='line 2')
ax.legend(['1', '2'], loc = 'lower right')
plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
line1, = ax.plot(np.arange(0, 5, 0.25), '--o', label='line 1')
line2, = ax.plot(np.arange(0, 10, 0.5),  ':s', label='line 2')
ax.legend((line2, line1), ['Линия 2', 'Линия 1'], bbox_to_anchor=(0.5, 0.7), facecolor='#aaa')
plt.show()
```

## Рисование геометрических фигур на графике

```python
from matplotlib.lines import Line2D
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
line1, = ax.plot(np.arange(0, 5, 0.25), '--o', label='line 1')
line2, = ax.plot(np.arange(0, 10, 0.5),  ':s', label='line 2')
ax.legend((line2, line1), ['Линия 2', 'Линия 1'], bbox_to_anchor=(0.5, 0.7), facecolor='#aaa')

l1 = Line2D([1, 2, 3], [1, 2, 4])
ax.add_line(l1)

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

l1 = Line2D([1, 2, 3], [1, 2, 4])
ax.add_line(l1)
ax.set(xlim=(1, 3), ylim=(1, 4))

plt.show()
```
