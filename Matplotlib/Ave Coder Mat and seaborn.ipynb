```python
figure, axes = plt.subplots(1, 2, figsize = (10, 5))

axes[0].hist(exam1_grades, color = 'g');
axes[1].hist(exam2_grades, color = 'b');
axes[0].set_title('Distribution of Grades for Exam #1')
axes[1].set_title('Distribution of Grades for Exam #2')

for a in axes.flat:
    a.set(xlabel='Grade(%)', ylabel='Count')

plt.show()
```

## Kernel Density Plots графики плотности ядра sns.kdeplot()

```python
scores = pd.Series(np.random.normal(50, 15, 400))
```

```python
mean = [60, 70]
covariance_matrix = [(1, .5), (.5, 1)]
```

```python
data = np.random.multivariate_normal(mean, covariance_matrix, size = 400)
```

```python
midterms = pd.DataFrame(data, columns=['midterm1', 'midterm2'])
```

```python
sns.kdeplot(scores, fill = True)
plt.show()
```

```python
midterms.head()
```

```python
sns.kdeplot(x = midterms['midterm1'], y = midterms['midterm2'], fill = True)
plt.show()
```
