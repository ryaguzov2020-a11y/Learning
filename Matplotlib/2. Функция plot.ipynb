```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
y = np.array([1, 2, -6, 0, 4]) # x - это индексы от 0 до 4
plt.plot(y)
plt.show()
```

```python
x = np.array([4, 5, 6, 7, 8])
y = np.array([1, 2, -6, 0, 4]) 
plt.plot(x, y)
plt.show()
```

```python
x = np.array([1, 1, 5, 5, 1])
y = np.array([1, 5, 5, 1, 1])
plt.plot(x, y)
```

```python
y = np.arange(5)             # [0, 1, 2, 3, 4]
x = np.array([a*a for a in y])   # [ 0,  1,  4,  9, 16]
y2 = [0, 1, 2, 3]
x2 = [i+1 for i in y2]
plt.plot(x, y, x2, y2)
plt.grid()
plt.show()
```

```python
y = np.arange(5)             # [0, 1, 2, 3, 4]
x = np.array([a*a for a in y])   # [ 0,  1,  4,  9, 16]
y2 = [0, 1, 2, 3]
x2 = [i+1 for i in y2]
plt.plot(x, y)
plt.plot(x2, y2)
plt.grid()
plt.show()
```

```python
y = np.arange(5)             # [0, 1, 2, 3, 4]
x = np.array([a*a for a in y])   # [ 0,  1,  4,  9, 16]
plt.plot(x, y, '--') 
plt.grid()
plt.show()
```

```python
y = np.arange(5)             # [0, 1, 2, 3, 4]
x = np.array([a*a for a in y])   # [ 0,  1,  4,  9, 16]
plt.plot(x, y, '--r', color = 'black') 
plt.grid()
plt.show()
```

```python
y = np.arange(5)             # [0, 1, 2, 3, 4]
x = np.array([a*a for a in y])   # [ 0,  1,  4,  9, 16]
plt.plot(x, y, 'r--o', marker= 's', markersize='10', markerfacecolor='w', linewidth='1') # o - маркер
plt.grid()
plt.show()
```

## Заливка областей графика

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)

plt.fill_between(x, y, 0.5, where=(y<0), color='r', alpha = 0.5)
plt.fill_between(x, y, 0.5, where=(y>=0), color='g', alpha = 0.5)

plt.grid()
plt.show()
```
