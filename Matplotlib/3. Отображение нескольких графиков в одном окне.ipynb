```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

subplot(nrows, ncols, index)

где nrows, ncols – число строк и столбцов; index – индекс текущих координатных осей (графика). Например, мы можем разбить все пространство окна на одну строку и три столбца, получим 1 x 3 = 3 поля для отображения графиков:

import numpy as np

```python
plt.subplot(1, 3, 1)
plt.plot(np.random.random(10))
plt.subplot(1, 3, 2)
plt.plot(np.random.random(10))
plt.subplot(1, 3, 3)
plt.plot(np.random.random(10))
plt.grid()
plt.show()
```

```python
ax1 = plt.subplot(1, 3, 1)
plt.plot(np.random.random(10))
ax2 = plt.subplot(1, 3, 2)
plt.plot(np.random.random(10))
ax3 = plt.subplot(1, 3, 3)
plt.plot(np.random.random(10))

ax1.grid()
ax2.grid()
ax3.grid()

plt.show()
```

```python
plt.subplot(2, 3, 1)
plt.plot(np.random.random(10))
plt.subplot(2, 3, 2)
plt.plot(np.random.random(10))
plt.subplot(2, 3, 3)
plt.plot(np.random.random(10))
plt.subplot(2, 1, 2)
plt.plot(np.random.random(10))
plt.grid()
plt.show()
```

## Функция subplots()

```python
f, ax = plt.subplots(2, 2)

f.set_size_inches(7, 4)
f.set_facecolor('grey')

ax[0, 0].plot(np.arange(0, 5, 0.2))
ax[0, 0].grid()
ax[0, 1].plot(np.arange(5, 0, -0.2))
ax[0, 1].grid()
plt.show()
```

## Объект Figure

```python
fig = plt.figure(figsize=(7, 4))
plt.plot(np.arange(0, 5, 0.2))
```

```python
fig = plt.figure(figsize=(5, 4))
ax1 = fig.add_axes([0.0, 0, 1.0, 1.0]) # [pos_x, pos_y, width, height]
ax1.plot(np.arange(0, 5, 0.2))
```

```python
fig = plt.figure(figsize=(7, 4))
ax1 = fig.add_subplot(1, 3, 1)
ax1.plot(np.arange(0, 5, 0.2))
```

## Компоновка графиков с помощью GridSpec

```python
from matplotlib.gridspec import GridSpec
```

```python
fig = plt.figure(figsize=(7, 4))

ws = [1, 2, 5]
hs = [2, 0.5]

gs = GridSpec(ncols=3, nrows=2, figure=fig, width_ratios=ws, height_ratios=hs)

ax1 = plt.subplot(gs[0, 0])
ax1.plot(np.arange(0, 5, 0.2))
ax2 = fig.add_subplot(gs[1, 0:2])
ax2.plot(np.random.random(10))
ax3 = fig.add_subplot(gs[:, 2])
ax3.plot(np.random.random(10))
```
